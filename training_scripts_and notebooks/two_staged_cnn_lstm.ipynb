```python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║   FINAL CNN-LSTM TRAINING SCRIPT                                     ║
# ║   Trains Stage 1 + Stage 2 with TemporalSumLayer                    ║
# ║   No Lambda layers — loads cleanly every time                        ║
# ║   Run all cells top to bottom in one Colab session                  ║
# ╚══════════════════════════════════════════════════════════════════════╝


# ─────────────────────────────────────────────────────────────────
# CELL 1 — Mount Drive & Imports
# ─────────────────────────────────────────────────────────────────
from google.colab import drive
import os, gzip, io
import numpy as np
import joblib
import matplotlib.pyplot as plt

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv1D, MaxPooling1D, LSTM, Dense,
    Dropout, Flatten, Activation, Multiply, Reshape
)
from tensorflow.keras.callbacks import (
    EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
)
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import (
    classification_report, confusion_matrix,
    mean_absolute_error, mean_squared_error
)

print("Mounting Google Drive...")
drive.mount('/content/drive')

PROJECT_FOLDER = '/content/drive/MyDrive/GNSS_csir'
os.chdir(PROJECT_FOLDER)
print(f"Working directory: {os.getcwd()}")

gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f"GPU: {gpus[0].name}")
    tf.config.experimental.set_memory_growth(gpus[0], True)

SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)


# ─────────────────────────────────────────────────────────────────
# CELL 2 — Configuration
# ─────────────────────────────────────────────────────────────────
X_PATH        = "X_sequences.npy"
Y_PATH        = "y_targets.npy"
TARGET_SCALER = "target_scaler.gz"
TEST_INDICES  = "test_indices.npy"

STAGE1_PATH   = "stage1_classifier.keras"
STAGE2_PATH   = "stage2_regressor.keras"

TIME_STEPS    = 48
N_FEATURES    = 13
BATCH_SIZE    = 256
EPOCHS        = 100
VAL_SPLIT     = 0.15
TEST_SPLIT    = 0.10


# ─────────────────────────────────────────────────────────────────
# CELL 3 — TemporalSumLayer (replaces Lambda permanently)
# ─────────────────────────────────────────────────────────────────
class TemporalSumLayer(tf.keras.layers.Layer):
    """Sums across time dimension — replaces Lambda layer."""
    def call(self, x):
        return tf.reduce_sum(x, axis=1)
    def get_config(self):
        return super().get_config()


# ─────────────────────────────────────────────────────────────────
# CELL 4 — Load Data & Create Splits
# ─────────────────────────────────────────────────────────────────
print("\n" + "="*55)
print("  STEP 1: Loading Data")
print("="*55)

def load_scaler(path):
    with gzip.open(path, 'rb') as f:
        return joblib.load(io.BytesIO(f.read()))

X             = np.load(X_PATH)
y             = np.load(Y_PATH)
target_scaler = load_scaler(TARGET_SCALER)

print(f"  X shape : {X.shape}")
print(f"  y shape : {y.shape}")
print(f"  Target max: {target_scaler.data_max_[0]:.4f} mm/hr ✓")

# Stage 1 labels
y_binary = (y.flatten() > 0).astype(np.float32)

# Stage 2 rain-only
rain_mask = y.flatten() > 0
X_rain    = X[rain_mask]
y_rain    = y[rain_mask]

# Chronological splits — Stage 1
test_idx   = np.load(TEST_INDICES)
test_start = int(test_idx[0])
val_start  = int(test_start * (1 - VAL_SPLIT))

X_tr,  y1_tr  = X[:val_start],           y_binary[:val_start]
X_vl,  y1_vl  = X[val_start:test_start], y_binary[val_start:test_start]
X_te,  y1_te  = X[test_start:],          y_binary[test_start:]

# Chronological splits — Stage 2
n2          = len(X_rain)
ts2         = int(n2 * (1 - TEST_SPLIT))
vs2         = int(ts2 * (1 - VAL_SPLIT))
X2_tr, y2_tr = X_rain[:vs2],    y_rain[:vs2]
X2_vl, y2_vl = X_rain[vs2:ts2], y_rain[vs2:ts2]
X2_te, y2_te = X_rain[ts2:],    y_rain[ts2:]

print(f"\n  Stage 1 — Train:{len(X_tr):,} Val:{len(X_vl):,} Test:{len(X_te):,}")
print(f"  Stage 2 — Train:{len(X2_tr):,} Val:{len(X2_vl):,} Test:{len(X2_te):,}")


# ─────────────────────────────────────────────────────────────────
# CELL 5 — Model Builders
# ─────────────────────────────────────────────────────────────────
def attention_block(x):
    """Attention using TemporalSumLayer — no Lambda."""
    attn    = Dense(1, activation='tanh')(x)
    attn    = Flatten()(attn)
    attn    = Activation('softmax')(attn)
    attn    = Reshape((x.shape[1], 1))(attn)
    context = Multiply()([x, attn])
    context = TemporalSumLayer()(context)
    return context


def build_stage1(time_steps=TIME_STEPS, n_features=N_FEATURES):
    inputs = Input(shape=(time_steps, n_features))
    x = Conv1D(64, 3, activation='relu', padding='same')(inputs)
    x = Conv1D(64, 3, activation='relu', padding='same')(x)
    x = MaxPooling1D(2)(x)
    x = Dropout(0.2)(x)
    x = LSTM(128, return_sequences=True)(x)
    x = Dropout(0.3)(x)
    x = LSTM(64, return_sequences=True)(x)
    x = Dropout(0.3)(x)
    x = attention_block(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.2)(x)
    x = Dense(32, activation='relu')(x)
    out = Dense(1, activation='sigmoid')(x)
    model = Model(inputs, out, name='Stage1_Classifier')
    model.compile(
        optimizer=Adam(1e-3),
        loss='binary_crossentropy',
        metrics=['accuracy',
                 tf.keras.metrics.AUC(name='auc'),
                 tf.keras.metrics.Precision(name='precision'),
                 tf.keras.metrics.Recall(name='recall')]
    )
    return model


def build_stage2(time_steps=TIME_STEPS, n_features=N_FEATURES):
    inputs = Input(shape=(time_steps, n_features))
    x = Conv1D(64, 3, activation='relu', padding='same')(inputs)
    x = Conv1D(64, 3, activation='relu', padding='same')(x)
    x = MaxPooling1D(2)(x)
    x = Dropout(0.2)(x)
    x = LSTM(128, return_sequences=True)(x)
    x = Dropout(0.3)(x)
    x = LSTM(64, return_sequences=True)(x)
    x = Dropout(0.3)(x)
    x = attention_block(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.2)(x)
    x = Dense(32, activation='relu')(x)
    out = Dense(1, activation='linear')(x)
    model = Model(inputs, out, name='Stage2_Regressor')
    model.compile(
        optimizer=Adam(1e-3),
        loss=tf.keras.losses.Huber(delta=1.0),
        metrics=['mae']
    )
    return model


def get_callbacks(save_path, patience=10):
    return [
        EarlyStopping(monitor='val_loss', patience=patience,
                      restore_best_weights=True, verbose=1),
        ModelCheckpoint(save_path, monitor='val_loss',
                        save_best_only=True, verbose=1),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                          patience=5, min_lr=1e-6, verbose=1)
    ]


# ─────────────────────────────────────────────────────────────────
# CELL 6 — Train Stage 1
# ─────────────────────────────────────────────────────────────────
print("\n" + "="*55)
print("  STAGE 1 — Rain/No-Rain Classifier")
print("="*55)

rain_ratio   = y_binary.mean()
class_weight = {0: 1.0, 1: round(float((1-rain_ratio)/rain_ratio), 4)}
print(f"  Class weights: {class_weight}")

tf.keras.backend.clear_session()
stage1 = build_stage1()
stage1.summary()

history1 = stage1.fit(
    X_tr, y1_tr,
    validation_data=(X_vl, y1_vl),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    class_weight=class_weight,
    callbacks=get_callbacks(STAGE1_PATH),
    verbose=1
)

# Evaluate
y1_prob = stage1.predict(X_te, batch_size=BATCH_SIZE, verbose=0).flatten()
y1_pred = (y1_prob > 0.5).astype(int)
print(classification_report(y1_te.astype(int), y1_pred,
                             target_names=['Dry', 'Rain']))
print(f"Confusion Matrix:\n{confusion_matrix(y1_te.astype(int), y1_pred)}")

# Verify it loads cleanly
reloaded_s1 = tf.keras.models.load_model(
    STAGE1_PATH,
    custom_objects={'TemporalSumLayer': TemporalSumLayer}
)
print(f"Stage 1 reload verified ✓  Saved → {STAGE1_PATH}")

# Plot
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history1.history['loss'],     label='Train')
plt.plot(history1.history['val_loss'], label='Val')
plt.title('Stage 1 — Loss'); plt.legend()
plt.subplot(1,2,2)
plt.plot(history1.history['accuracy'],     label='Train')
plt.plot(history1.history['val_accuracy'], label='Val')
plt.title('Stage 1 — Accuracy'); plt.legend()
plt.tight_layout()
plt.savefig('stage1_curves.png', dpi=150)
plt.show()


# ─────────────────────────────────────────────────────────────────
# CELL 7 — Train Stage 2
# ─────────────────────────────────────────────────────────────────
print("\n" + "="*55)
print("  STAGE 2 — Rainfall Amount Regressor")
print("="*55)
print(f"  Training on {len(X2_tr):,} rainy samples only")

stage2 = build_stage2()
stage2.summary()

history2 = stage2.fit(
    X2_tr, y2_tr,
    validation_data=(X2_vl, y2_vl),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=get_callbacks(STAGE2_PATH),
    verbose=1
)

# Evaluate Stage 2
y2_pred_s  = stage2.predict(X2_te, batch_size=BATCH_SIZE, verbose=0)
y2_pred_s  = np.clip(y2_pred_s, 0, 1)
y2_pred_mm = np.clip(target_scaler.inverse_transform(y2_pred_s), 0, None)
y2_true_mm = np.clip(target_scaler.inverse_transform(y2_te), 0, None)

mae2  = mean_absolute_error(y2_true_mm, y2_pred_mm)
rmse2 = np.sqrt(mean_squared_error(y2_true_mm, y2_pred_mm))
print(f"\n  Stage 2 MAE  : {mae2:.4f} mm/hr")
print(f"  Stage 2 RMSE : {rmse2:.4f} mm/hr")

# Verify it loads cleanly
reloaded_s2 = tf.keras.models.load_model(
    STAGE2_PATH,
    custom_objects={'TemporalSumLayer': TemporalSumLayer}
)
print(f"Stage 2 reload verified ✓  Saved → {STAGE2_PATH}")

# Plot
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history2.history['loss'],     label='Train')
plt.plot(history2.history['val_loss'], label='Val')
plt.title('Stage 2 — Loss'); plt.legend()
plt.subplot(1,2,2)
plt.plot(history2.history['mae'],     label='Train')
plt.plot(history2.history['val_mae'], label='Val')
plt.title('Stage 2 — MAE'); plt.legend()
plt.tight_layout()
plt.savefig('stage2_curves.png', dpi=150)
plt.show()


# ─────────────────────────────────────────────────────────────────
# CELL 8 — Combined Two-Stage Inference
# ─────────────────────────────────────────────────────────────────
print("\n" + "="*55)
print("  COMBINED TWO-STAGE INFERENCE ON TEST SET")
print("="*55)

prob_rain    = stage1.predict(X_te, batch_size=BATCH_SIZE, verbose=0).flatten()
rain_flags   = prob_rain > 0.5

final_scaled = np.zeros((len(X_te), 1), dtype=np.float32)
if rain_flags.sum() > 0:
    final_scaled[rain_flags] = stage2.predict(
        X_te[rain_flags], batch_size=BATCH_SIZE, verbose=0
    )

final_scaled  = np.clip(final_scaled, 0, 1)
final_pred_mm = np.clip(
    target_scaler.inverse_transform(final_scaled), 0, None
).flatten()
final_true_mm = np.clip(
    target_scaler.inverse_transform(y[test_start:]), 0, None
).flatten()

mae_combined  = mean_absolute_error(final_true_mm, final_pred_mm)
rmse_combined = np.sqrt(mean_squared_error(final_true_mm, final_pred_mm))
print(f"  Combined MAE  : {mae_combined:.4f} mm/hr")
print(f"  Combined RMSE : {rmse_combined:.4f} mm/hr")

# Save test predictions for ensemble script
np.save('cnn_lstm_test_pred_scaled.npy', final_scaled)
print(f"  Saved → cnn_lstm_test_pred_scaled.npy")

plt.figure(figsize=(14,4))
plt.plot(final_true_mm[:500], label='Actual',    alpha=0.75)
plt.plot(final_pred_mm[:500], label='Predicted', alpha=0.75)
plt.title(f'CNN-LSTM — Predictions vs Actual | MAE={mae_combined:.4f} mm/hr')
plt.xlabel('Sample Hour'); plt.ylabel('Precipitation (mm/hr)')
plt.legend(); plt.tight_layout()
plt.savefig('cnn_lstm_final_predictions.png', dpi=150)
plt.show()

print(f"""
  ════════════════════════════════════════
  CNN-LSTM TRAINING COMPLETE
  ════════════════════════════════════════
  Saved:
    stage1_classifier.keras   ✓
    stage2_regressor.keras    ✓
    cnn_lstm_test_pred_scaled.npy ✓

  Both models verified loadable with:
    custom_objects={{'TemporalSumLayer': TemporalSumLayer}}

  Next: Run script_d_ensemble.py
  in a fresh Colab session
  ════════════════════════════════════════
""")
```
